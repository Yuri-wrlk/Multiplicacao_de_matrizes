```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

mat_mult = pd.read_csv("data/Results.csv", sep=",")
mat_mult.head()
```

```python
conc = list()
sequential = list()

conc = mat_mult[mat_mult.execution_type == 'C']
sequential = mat_mult[mat_mult.execution_type == 'S']

x_conc_heights = conc[conc.columns[2]]
x_conc_time = conc[conc.columns[3]]

x_seq_heights = sequential[sequential.columns[2]]
x_seq_time = sequential[sequential.columns[3]]

conc[conc.columns[2]]
```

```python
sns.set_style("darkgrid")
plt.suptitle("Tempo de execução concorrente", 
             fontsize=20,
            y = 1.09)
plt.xticks(rotation=60)
plt.yticks(rotation=60)
plt.xlabel('Tempo (em segundos)')
plt.ylabel('Nº de dimensões')
plt.xlim(-5, 1100)
plt.ylim(-5, 600)


plt.plot([x_conc_heights , x_seq_heights], [x_conc_time. x_seq_time])
plt.show()
```

```python
sns.set_style("darkgrid")
plt.suptitle("Tempo de execução sequencial", 
             fontsize=20,
            y = 1.09)
plt.xticks(rotation=60)
plt.yticks(rotation=60)
plt.xlabel('Tempo (em segundos)')
plt.ylabel('Nº de dimensões')
plt.xlim(-5, 1100)
plt.ylim(-5, 600)



plt.plot(x_seq_heights, x_seq_time)
plt.show()
```
